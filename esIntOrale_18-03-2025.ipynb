```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("./csv/train_energy_data.csv")
data.head()
```

```python
data.info()
```

## Da fare:
Regressione lineare di predire il consumo energetico di un edificio a partire dalle caratteristiche a disposizione

```python
encoded = LabelEncoder()
data["Building Type"] = LabelEncoder().fit_transform(data["Building Type"])
data["Day of Week"] = LabelEncoder().fit_transform(data["Day of Week"])
```

```python
target = ["Energy Consumption"]
features = ["Building Type", "Square Footage", "Number of Occupants", "Appliances Used", "Average Temperature", "Day of Week"]
```

```python
X = data[features]
y = data[target]
```

divido in train e test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)
```

scelgo modello e alleno

```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)
```

valuto

```python
model.score(X_test, y_test)
```

predico

```python
consumoEnergetico = model.predict(X_test)
consumoEnergetico
```

salvo notebook

```python
import joblib

joblib.dump(model, 'energy_consumption_model.joblib')
joblib.dump(LabelEncoder(), 'building_type_encoder.joblib')
joblib.dump(LabelEncoder(), 'day_of_week_encoder.joblib')
```
