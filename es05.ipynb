```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim = 8
chessboard = np.zeros((dim, dim), dtype=np.int32)
#slicing(assegnazione con broadcast)
chessboard[::2, ::2] = 1 #salta di due righe, salta le celle a due a due (fa le righe dispari)
chessboard[1::2, 1::2] = 1 #idem ma per le righe pari
chessboard
```

```python
plt.imshow(chessboard, cmap="Greys")
```

```python
dim = 1000
array2d = np.zeros((dim, dim), dtype=np.float32)
array2d[:] = np.linspace(0., 1., dim)
plt.imshow(array2d, cmap="magma") #Greys, hsv, inferno, gnuplots, jet, rainbow
```

### Creo un bersaglio di quadrati concentrici (si può usare al massimo un ciclo for).
### BONUS: ogni cornice del bersaglio ha un colore diverso

```python
l = 50 #aumentare la dimensione per vedere il cambiamento del gradiente dell'img
array2d = np.zeros((l, l), dtype = np.float32)

for i in range(0, l +1):
    array2d[i: -i, i : - i] = i
    
plt.imshow(array2d, cmap="magma")
```
