Il file `Netflix-Dataset-Latest-2021.csv` contiene svariati dati di film e serie Netflix.

1) Utilizza matplotlib creare un grafico a dispersione che mostri la correlazione tra `IMDb Score` e `Awards Received`. I punti del grafico devono avere un colore che dipenda dal numero di voti ricevuti `IMDb Votes`. I grafici devono avere titolo, griglia ed etichette sugli assi.

2) Traccia lo stesso grafico del punto 1 facendo in modo che la dimensione dei punti sia proporzionale al numero di voti `IMDb Votes`.

3) Traccia in un'unica figura due due istogrammi che rappresentino rispettivamente le distribuzioni di `IMDb Score` e `Awards Received`.

```python
dati = pd.read_csv("./Netflix-Dataset-Latest-2021.csv")
dati.head()
```

```python
dati.info()
```

```python
#1 - Grafico a dispersione: Colore basato su IMDb Votes
plt.figure(figsize = (10, 6))
plt.scatter(dati['IMDb Score'], dati['Awards Received'], c = dati['IMDb Votes'], alpha = 0.7, edgecolor = 'k')
plt.title('Grafico a dispersione - correlazione IMDb score con awards (Colore: IMDb Votes)')
plt.xlabel('Punteggi IMDb')
plt.ylabel('Premi ricevuti')
plt.grid()
plt.show()
```

```python
#2 - Grafico a dispersione: dimensione basato su IMDb Votes
plt.figure(figsize = (5, 6))
plt.scatter(dati['IMDb Score'], dati['Awards Received'], c = dati['IMDb Votes'], alpha = 0.7, s = dati['IMDb Votes']/1000, edgecolor = 'k') #/1000 per leggibilità sennò veniva gigante e non si capiva niente
plt.title('Correlazione tra IMDb Score e Awards Received (Dimensione: IMDb Votes)')
plt.xlabel('Punteggi IMDb')
plt.ylabel('Premi ricevuti')
plt.grid()
plt.show()
```

```python
#3 - distribuzioni di IMDb Score e Awards Received
plt.figure(figsize = (15, 6))

#IMDb Score
plt.subplot(1, 2, 1)
plt.hist(dati['IMDb Score'], bins = 15, color = 'violet', edgecolor = 'k')
plt.title('Istogramma di IMDb Score')
plt.xlabel('Punteggi IMDb')
plt.ylabel('Frequenza')
plt.grid()

#Awards Received
plt.subplot(1, 2, 2)
plt.hist(dati['Awards Received'], bins = 15, color = 'orange', edgecolor = 'k')
plt.title('Istogramma di Awards Received')
plt.xlabel('Premi ricevuti')
plt.ylabel('Frequenza')
plt.grid()

plt.show()
```
