### Esercizio meteo (pag.135 es11)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dati = pd.read_csv("./MonthElb_012010_004028900.csv", header = None, na_values = 9999.9)
dati.head()
```

```python
dati.info()
```

```python
colsName = {5:"VELV", 8:"BARO", 9:"TERMA", 10:"IGRO", 13:"PLUV"}
cols = [5, 8, 9, 10, 13]
dati[cols]
```

```python
fig, axs = plt.subplots(nrows = 5, ncols = 5, figsize = (25, 25))
for i, icol in enumerate(cols):
    for j, jcol in enumerate(cols):
        if i != j:
            x = dati[icol].values
            y = dati[jcol].values
            axs[i, j].scatter(x, y, s = 2, c = "#FE5D26")
            axs[i, j].set_xlabel(colsName[icol])
            axs[i, j].set_ylabel(colsName[jcol])
            axs[i, j].set_facecolor("#FAEDCA")
        else:
            x = dati[icol].values #x è = a y
            axs[i, j].hist(x, color = "#FE5D26", edgecolor = "#F2C078")
            axs[i, j].set_xlabel(colsName[icol])
            axs[i, j].set_ylabel("Frequenza assoluta")
            axs[i, j].set_facecolor("#FAEDCA")
fig.suptitle("Analisi grafica meteorologica", fontsize = 40, y = 0.92)
plt.show()
```

### fare i grafici a linea che mostrino nel tempo l'andamento delle pioggi PLUV e della temperatura TERMA

```python
colsName = {1: "HOURS", 9:"TERMA", 13:"PLUV", 14:"DATE"}
dati[14] = pd.to_datetime(dati[0] + " " +dati[1], format = "%d-%m-%Y %H:%M")
cols = [9, 13, 14]
dati[cols]
dati[cols].head()
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (25, 5))
axs[0].plot(dati[14], dati[13], c = "#FE5D26")
axs[1].plot(dati[14], dati[9], c = "#FE2D26")
axs[0].set_xlabel(colsName[14])
axs[0].set_ylabel(colsName[13])
axs[1].set_xlabel(colsName[14])
axs[1].set_ylabel(colsName[9])
fig.suptitle("Analisi grafica meteorologica temporale", fontsize = 40, y = 0.98)
plt.show()
```
