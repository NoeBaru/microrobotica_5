grafici su matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

esempio di grafico a linea

```python
x = np.linspace(-5, 5, 1000)
y = np.exp(-x**2)

x_t = np.linspace(-7, 7, 1000)
y_t = np.exp(-x_t**2)

fig, ax = plt.subplots(figsize =( 6, 4))
fig.set_facecolor("white")
ax.set_facecolor("#F5F5DC")
ax.plot(x_t, y_t, linestyle="--", linewidth = 2, color = "#CE5E6F", label=None)
ax.plot(x, y, linewidth = 2, color = "#CE5E6F", label = "$e^{-x^2}$") #oppure color = np.array([0.5, 0.1, 0.4, 0.4])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Il grafico di una curva Gaussiana", fontsize = 10)
ax.legend()
ax.grid()
plt.show()
```

```python
nPunti = 100
x = np.random.uniform(0, 1, nPunti)
y = np.random.uniform(0, 1, nPunti)

x_1 = np.random.uniform(0, 1, nPunti)
y_1 = np.random.uniform(0, 1, nPunti)

fig, ax = plt.subplots(figsize = (5, 5))
ax.set_facecolor("#1B1463")
ax.plot(x, y, linestyle = " ", marker = "o", color = "#F7F2EA", markersize = 5) #conta l'ordine di scrittura, se metto la riga dopo, i pallini vanno in secondo piano
ax.plot(x_1, y_1, linestyle = " ", marker = "X", color = "#4865ff", markersize = 5)

plt.show()
```

```python
nDots = 100
x = np.random.uniform(0, 1, nDots)
y = np.random.uniform(0, 1, nDots)
size = np.random.uniform(2, 20, nDots)
color = np.random.choice(["white" ,"yellow"], nDots)
```

```python
fig, ax = plt.subplots(figsize = (3,3))
fig.set_facecolor("#1282A2")
ax.set_facecolor("#0A1128")
ax.scatter(x, y, c = color, s = size, marker = "*", alpha = 0.5) #ogni punto pu essere diverso, con plot ogni punto deve essere uguale
plt.show()
```

```python
nDots = 100
x1 = np.random.uniform(0, 1, nDots)
y1 = np.random.uniform(0, 1, nDots)
x2 = np.random.uniform(0, 1, nDots)
y2 = np.random.uniform(0, 1, nDots)
x3 = np.random.uniform(0, 1, nDots)
y3 = np.random.uniform(0, 1, nDots)
size = np.random.uniform(100, 500, nDots)
```

```python
fig, ax = plt.subplots(figsize = (5,5))
#fig.set_facecolor("white")
# ax.set_facecolor("#0A1128")
ax.scatter(x1, y1, label = "tab:blue",  c = "blue", s = size, alpha = 0.5, linewidth = 0) #ogni punto pu essere diverso, con plot ogni punto deve essere uguale
ax.scatter(x2, y2, label = "tab:orange", c = "orange", s = size, alpha = 0.5, linewidth = 0)
ax.scatter(x3, y3, label = "tab:green", c = "green", s = size, alpha = 0.5, linewidth = 0)
ax.legend(loc="upper right")
ax.grid()
plt.show()
```

### pandas

```python
import pandas as pd
```

```python
data = pd.read_csv("./penguins_size.csv")
data.head()
```

```python
data.dropna(inplace = True) #non prende i valori nulli
```

```python
bodyMassAdelie = data[data["species"]=="Adelie"]["body_mass_g"]
bodyMassAdelie
```

```python
bodyMassGentoo = data[data["species"]=="Gentoo"]["body_mass_g"]
bodyMassGentoo
```

```python
bins = np.linspace(2500, 6500, 20) #intervalli personalizzati per l'istogramma
fig, ax = plt.subplots(figsize = (4,3))
ax.set_facecolor("#E8DAB2")
ax.hist(bodyMassAdelie, label = "Adelie", edgecolor = "k", alpha = 0.7, bins = bins, color = "#DD6E42") #colore del colore e k significa black
ax.hist(bodyMassGentoo, label = "Gentoo", edgecolor = "k", alpha = 0.7, bins = bins, color = "#4F6D7A")
ax.set_xlabel("Peso corporeo (g)")
ax.set_ylabel("N esemplari")
ax.legend()
ax.grid()
plt.savefig("./istogrammaAdelieGentoo.png", dpi = 200, bbox_inches = "tight")
plt.show()
```

prendo un'immmagine, faccio due grafici (sx l'immagine, in dx tre istogramma che mostri la distibuzione dell'rgb)

```python
image = plt.imread("./netflix_crop_resize.jpg") #le immagini sono array tridimensionali
image.shape
plt.imshow(image)
```

```python
red = image[:,:,0]
green = image[:,:, 1]
blue = image[:,:, 2]
image_copy = image.copy()
```

```python
red = np.reshape(red, -1)
green = np.reshape(green, -1)
blue =np.reshape(blue, -1)
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 4))
axs[0].imshow(image)
axs[1].hist(red, color="red", alpha=0.3, label="Red")
axs[1].hist(green, color="green", alpha=0.3, label="Green")
axs[1].hist(blue, color="blue", alpha=0.3, label="Blue")
axs[1].set_title("RGB Histogram")
axs[1].set_xlabel("Pixel Intensity")
axs[1].set_ylabel("Frequency")
axs[1].legend()
plt.show()
```
