creare un'altra funzione come nell'es02 e creo un dizionario di liste del carattere successivo

```python
import random
```

```python
def bigramsNCounter(fileName, n): #funzione dell'es02 modificata
    bigrams = {}
    
    with open(fileName, 'r') as file:
            text = file.read()
    
            for i in range(len(text) - n):
                bigram = text[i:i + n]  #per estrarre il bigramma
                nextChar = text[i + n] if i + n < len(text) else None  # carattere successivo
    
                if bigram not in bigrams:
                    bigrams[bigram] = []
                if nextChar:
                    bigrams[bigram].append(nextChar)  # aggiunge il carattere successivo
    return bigrams
```

```python
#main
fileName = 'OSCAR_WILDE_Il_fantasma_di_Canterville.txt'
n = int(input("inserisci un numero: "))
results = bigramsNCounter(fileName, n)

for bigram, nextChars in results.items():
    print(f"{bigram}: {nextChars}")
```
