### Esercizio 1
Sorteggiare 25 cifre a caso e rappresentarle graficamente in una figura. Indicare nel titolo di ciascuna la cifra rappresentata

```python
dict = {}
for c in range(0, 10):
    dict[c] = digits[digitsNames == c]
dict
```

```python
fig, axs = plt.subplots(nrows = 10, ncols = 10, figsize = (4, 4))
for c in range(0, 10): #con -1 se lo calcola lui
    digitsSel = dict[c]
    num = digitsSel.shape[0]
    i = np.random.choice(np.arange(0, num), 10, replace = False) #replace false vuol dire che una volta sorteggiato non può riuscire (come da un sacchetto)
    digitsSel = digitsSel[i]
    for col, dig in enumerate(digitsSel):
        axs[c, col].imshow(dig.reshape((28,28)), cmap = "Greys")
        axs[c,col].set_axis_off()
plt.show()
```
