```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('./csv/GlobalWeatherRepository.csv', parse_dates = ["last_updated"])
data.head()
```

```python
data.info()
```

```python
#a partire dal 1/01/1970, anno 0 degli informatici

andorra = data[data["country"]=="Andorra"]
andorra
```

```python
cols = ["last_updated", "temperature_celsius", "wind_kph", "precip_mm", "humidity" , "cloud"]
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(andorra["last_updated"], andorra["temperature_celsius"])
plt.show()
```

```python
fig, axs = plt.subplots(figsize = (10, 4))
for i,n in cols:
    axs[i].plot(cols[0], cols[n+1])
    axs[i].grid()
plt.show()
```
