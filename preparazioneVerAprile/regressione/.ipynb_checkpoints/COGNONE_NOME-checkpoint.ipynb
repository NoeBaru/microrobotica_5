Una compagnia di assicurazioni vuole creare un algoritmo per stimare le spese sanitarie dei clienti a partire da alcuni dati rilevanti.
A tal fine devi creare una regressione lineare usando i dati presenti nel file `insurance.csv`. Il file contiene le seguenti colonne:
* `age`: età del clienti;
* `sex`: il sesso dei clienti;
* `bmi`: l'indice di massa grassa dei clienti;
* `children`: il numero di figli dei clienti;
* `smoker` se fumatore oppure no;
* `region` la regione di residenza;
* `charges` il valore delle spese sanitarie sostenute in dollari.

1. Realizza la regressione lineare e valutane il coefficiente di determinazione. 
2. Seleziona le feature che massimizzano il coefficiente di determinazione della regressione.
3. Per ogni coppia di feature crea un grafico in cui rappresenti i dati e la retta di regressione.


Note docente, da cancellare prima della somministrazione

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
data = pd.read_csv("./insurance.csv")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
le_sex = LabelEncoder()
le_sex.fit(data["sex"])
data["sex_e"]=le_sex.transform(data["sex"])

le_smoker = LabelEncoder()
le_smoker.fit(data["smoker"])
data["smoker_e"]=le_smoker.transform(data["smoker"])

le_region = LabelEncoder()
le_region.fit(data["region"])
data["region_e"]=le_region.transform(data["region"])

data
```

```python
le_region.classes_
```

```python
#estrazione dei dati in array numpy
X = data[['age',"bmi","sex_e","region_e","smoker_e"]].values
y = data[['charges']].values
#standard scaler per la feature
sc_x = StandardScaler()
#standard scaler per la variabile target
sc_y = StandardScaler()
#normalizza le feature
X_sc = sc_x.fit_transform(X)
#normalizza la variabile target
y_sc = sc_y.fit_transform(y)
```

```python
#creazione e addestramento della regressione lineare
reg = LinearRegression().fit(X_sc, y_sc)

print (f"Il coefficiente di determinazione R^2 vale: {reg.score(X_sc, y_sc):.2f}")
print (f"Il coefficiente angolare vale: {reg.coef_[0][0]:.2f}")
print (f"L'intercetta vale: {reg.intercept_[0]:.2f}")
```

```python
fig, ax = plt.subplots(figsize=(18, 12))
ax.tick_params(axis='both', labelsize=14)
ax.set_title('titolo', fontsize=22)
data.plot.scatter(x='bmi', y='charges',
                       label='dati misurati', s=10.,color="r",ax=ax)
#data["Linear Regression"].plot(label='regressione lineare', lw =4, ax=ax, color ='b')
ax.set_ylabel('x', fontsize=20)
ax.set_xlabel('y', fontsize=20)

plt.legend(fontsize=20)
plt.show()
```
