Rappresentare su un grafico a barre la distribuzione dei fumatori maschi per fascia dâ€™etÃ  ( 18-30 / 30-45/ 45-60 / 60-80 / 80 e oltre).
Specificare nome degli assi, titolo e dimensione  del grafico, colore


```python
#filtro per genere maschile e fumatori
maleSmokers = data[(data['Gender'] == 'Male') & (data['Smoking'] == 'Yes')]

#creo fasce d'etÃ 
bins = [18, 30, 45, 60, 80, 120]
labels = ['18-30', '31-45', '46-60', '60-80', '80+']
maleSmokers['Age_Group'] = pd.cut(maleSmokers['Age'], bins=bins, labels=labels, right=False)

#conto il numero di fumatori per fascia d'etÃ 
ageGroupCounts = maleSmokers['Age_Group'].value_counts().reindex(labels)

#creo il grafico a barre
plt.figure(figsize=(10, 6))
ageGroupCounts.plot(kind='bar', color='blue', alpha=0.7, edgecolor='black')

#aggiungo titoli ed etichette
plt.title("Numero di fumatori (maschi) per fasce d'etÃ ", fontsize=16)
plt.xlabel("Fasce di etÃ ", fontsize=14)
plt.ylabel("Numero di fumatori", fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

#mostro il grafico
plt.tight_layout()
plt.show()
```
