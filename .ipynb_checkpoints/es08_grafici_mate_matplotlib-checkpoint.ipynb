* la posizione di dove si trova il massimo (m valore medio)
* la larghezza (s(sigma) deviazione standard).

Se il punteggio Ã¨ maggiore pt+1 se pari pt+0.5 a testa sennÃ² pt0
Simulare 10 partite da 1000 l'una

```python
dado = [1, 2, 3, 4, 5, 6]
nPartite = 10

g1 = np.random.choice(dado, (nPartite, 1000))
g2 = np.random.choice(dado, (nPartite, 1000))

vinceG1 = g1 > g2
pareggio = g1 == g2
vinceG2 = g1 < g2

ptG1 = np.count_nonzero(vinceG1) + 0.5*np.count_nonzero(pareggio)
print(f"Puti vittorie g1 {ptG1}")

ptG2 = np.count_nonzero(vinceG2) + 0.5*np.count_nonzero(pareggio)
print(f"Punti vittorie g2 {ptG2}")
```

Se un giocatore usa un dado truccato, in cui il 5 e il 6 hanno probabilitÃ  doppia di uscire rispetto agli altri numeri

```python
dado = [1, 2, 3, 4, 5, 6]
nPartite = 10
probabilita = 1/8

g1 = np.random.choice(dado, (nPartite, 1000), p = probabilita)
g2 = np.random.choice(dado, (nPartite, 1000))

vinceG1 = g1 > g2
pareggio = g1 == g2
vinceG2 = g1 < g2

ptG1 = np.count_nonzero(vinceG1) + 0.5*np.count_nonzero(pareggio)
print(f"Puti vittorie g1 {ptG1}")

ptG2 = np.count_nonzero(vinceG2) + 0.5*np.count_nonzero(pareggio)
print(f"Punti vittorie g2 {ptG2}")
```
