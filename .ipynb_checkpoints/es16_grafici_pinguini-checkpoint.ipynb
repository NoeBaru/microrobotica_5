### grafici pinguini

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dati = pd.read_csv("csv/penguins_size.csv")
dati.head()
```

```python
dati.dropna(inplace = True) #rimuove valori mancanti
dati.head()
```

```python
dati.info()
```

```python
cols = ["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]
dati[cols].head() #prende le prime 5 in automatico, se metto (n) farà vedere n righe
```

```python
fig, axs = plt.subplots(nrows = 4, ncols = 4, figsize = (20, 20))
for i, icol in enumerate(cols):
    for j, jcol in enumerate(cols):
        if i != j:
            x = dati[icol].values
            y = dati[jcol].values
            axs[i, j].scatter(x, y, s = 2, c = "#FE5D26")
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel(jcol)
            axs[i, j].set_facecolor("#FAEDCA")
        else:
            x = dati[icol].values #x è = a y
            axs[i, j].hist(x, color = "#FE5D26", edgecolor = "#F2C078")
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel("Frequenza assoluta")
            axs[i, j].set_facecolor("#FAEDCA")
plt.show()
```

```python
pengColors = ["#C5D86D", "#86615C", "#AFE0CE"]
spc = dati["species"].values
spcColor = spc.copy()
spcColor[spcColor == "Adelie"] = pengColors[0]
spcColor[spcColor == "Chinstrap"] = pengColors[1]
spcColor[spcColor == "Gentoo"] = pengColors[2]
```

```python
sfondo = "#FFEDDF"
fig, axs = plt.subplots(nrows = 4, ncols = 4, figsize = (20, 20))
for i, icol in enumerate(cols):
    for j, jcol in enumerate(cols):
        if i != j:
            x = dati[icol].values
            y = dati[jcol].values
            axs[i, j].scatter(x, y, s = 15, c = spcColor)
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel(jcol)
            axs[i, j].set_facecolor("#FAEDCA")
        else:
            x = dati[icol].values #x è = a y
            axs[i, j].hist(x, color = "#FE5D26", edgecolor = "#F2C078")
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel("Frequenza assoluta")
            axs[i, j].set_facecolor("#FAEDCA")
plt.show()
```

### Esercizio 1
Rifare l'esempio di prima con il file marziani.csv (senza usare la colonna dei colori)

```python
marziani = pd.read_csv("./marziani.csv")
marziani.head()
```

```python
mCols = ["specie", "n_arti", "peso", "altezza", "larghezza"]
marziani[mCols].head() #prende le prime 5 in automatico, se metto (n) farà vedere n righe
```

```python
marzColors = ["purple", "pink"]
spc = marziani["specie"].values
spcColor = spc.copy()
spcColor[spcColor == "Robby"] = marzColors[0]
spcColor[spcColor == "Simmy"] = marzColors[1]
```

```python
sfondo = "white"
fig, axs = plt.subplots(nrows = 5, ncols = 5, figsize = (20, 20))
for i, icol in enumerate(mCols):
    for j, jcol in enumerate(mCols):
        if i != j:
            x = marziani[icol].values
            y = marziani[jcol].values
            axs[i, j].scatter(x, y, s = 2, c = spcColor)
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel(jcol)
            axs[i, j].set_facecolor("white")
        else:
            x = marziani[icol].values #x è = a y
            axs[i, j].hist(x, color = "blue", edgecolor = "black")
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel("Frequenza assoluta")
            axs[i, j].set_facecolor("white")
plt.show()
```
