```python
#selezionare tutte le auto di marca BMW
data[ data["MAKE"]== "BMW" ]
```

```python
#selezionare tutte le auto di marca BMW che hanno 10 cilindri
data[ (data["MAKE"]== "BMW") & (data["CYLINDERS"] == 10)]
```

```python
#selezionare tra tutte le auto di marca BMW quante sono per cilindri
data[data["MAKE"]== "BMW"]["CYLINDERS"].value_counts()
```

```python
#x o z benzina in base agli ottani, d diesel
data[["FUEL", "EMISSIONS"]].groupby("FUEL").mean()
```

```python
# per una marca conteggio le diverse tipologie di veicolo per quella macchina
data[data["MAKE"]== "BMW"]["VEHICLE CLASS"].value_counts()
```

```python
# in media, a parità di numeri di cilindri, consuma di più in media una BMW a 4 cilindri o un'AUDI a 4 cilindri?
audiBmw = (data["MAKE"] == "BMW") | (data["MAKE"]== "AUDI")
cilindri4 = data["CYLINDERS"] == 4
data[audiBmw & cilindri4][["MAKE", "EMISSIONS"]].groupby("MAKE").mean()
```

```python
data[audiBmw][["MAKE", "CYLINDERS", "EMISSIONS"]].groupby(["MAKE", "CYLINDERS"]).mean() #media auto consumi per cilindri
```

```python
data[audiBmw][["MAKE", "CYLINDERS", "EMISSIONS"]].groupby(["MAKE", "CYLINDERS"]).count() #quante auto per cilindri
```

```python
data[audiBmw][["MAKE", "CYLINDERS", "EMISSIONS"]].groupby(["MAKE", "CYLINDERS"]).min() #emissioni minime
```

```python
data[audiBmw][["MAKE", "CYLINDERS", "EMISSIONS"]].groupby(["MAKE", "CYLINDERS"]).max() #emissioni massime
```

```python
# trovare il consumo medio per marca considerando tutte le auto
mediaConsumi = data[["MAKE", "EMISSIONS"]].groupby("MAKE").mean()
```
