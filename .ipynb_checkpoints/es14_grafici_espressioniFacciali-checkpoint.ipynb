```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dati = pd.read_csv("./fer2013.csv")
```

```python
dictEmotions = {0: "Angry", 1: "Disgust", 2: "Fear", 3: "Happy", 4: "Sad", 5: "Surprise", 6: "Neutral"}
```

```python
dati.head(10)
```

```python
emotions = dati["emotion"].values #converte una colonna di un dataframe in un array
emotions
```

```python
expressions = dati["pixels"].values #array di stringhe
expressions.shape #quanti elementi ci sono dentro
```

```python
idFaccia = 4747
faccia1d = np.array([int(n) for n in expressions[idFaccia].split(" ")])
faccia1d
```

```python
faccia1d.shape
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.imshow(faccia1d.reshape((48, 48)), cmap="Greys_r") #con _r si inverte il colore
ax.set_axis_off()
plt.show()
```

```python
emotion = dictEmotions[emotions[idFaccia]]
emotion
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.imshow(faccia1d.reshape((48, 48)), cmap="Greys_r") #con _r si inverte il colore
ax.set_title(dictEmotions[emotions[idFaccia]])
ax.set_axis_off()
plt.show()
```

### Esercizio 1
Creare una figura con 25 espressioni facciali sorteggiate a caso. Per ognuna indicare quale emozione nel titolo.

```python
def string2face(s):
    faccia1d = np.array([int(n) for n in s.split(" ")])
    return faccia1d.reshape((48, 48))
```

```python
ids = np.random.choice(np.arange(0, len(expressions)), 25, replace = False)
ids
```

```python
emotions[ids]
```

```python
choosenEmotions = emotions[ids]
choosenExpressions = expressions[ids]
```

```python
fig, axs = plt.subplots(nrows = 5, ncols = 5, figsize = (6, 8))
for i, ax in enumerate(axs.reshape(-1)): #con -1 se lo calcola lui
    ax.imshow(string2face(choosenExpressions[i]), cmap = "Greys_r")
    ax.set_title(dictEmotions[choosenEmotions[i]], fontsize = 8)
    ax.set_axis_off()
plt.show()
```
