```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("csv/smartphones.csv")
data.head()
```

```python
#tutti i prezzi di tutti gli apple e tutti i samsung e faccio un grafico a distribuzione vicino all'altro
data.info()
```

```python
apple = data[data['phone_brand'] == "apple"]
samsung = data[data['phone_brand'] == "samsung"]
applePrices = apple['price_usd']
samsungPrices = samsung['price_usd']

fig, ax = plt.subplot(figsize = (4,3))
ax.hist(applePrices, color = 'violet', bins = 30, edgecolor = 'k', label =  "apple")
ax.hist(samsungPrices, color = 'orange', bins = 30, edgecolor = 'k', label = "samsung")
ax.set_title('Istogramma dei prezzi Apple')
ax.set_xlabel('Marca')
ax.set_ylabel('Prezzo')
ax.legend()
ax.grid()
plt.show()
```
