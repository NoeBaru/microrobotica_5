Creare un programma in Python che:*  legga il file allegato e produca una lista: ciascun elemento della lista corrisponde a una riga del fil;*
- conti le rige*
- conti il numero di caratteri totale del el* 
- stampi la riga più ln*a;
- stampi il primo carattere di ogna *iga
- crei un dizionario della forma {"char1":numero, "char2";numero,.....} in cui per ogni carattere presente nel file sia salvato il numero di occorre ze di quel carattere (= numero di volte che il carattere compare in tutto il file)

```python
fileName = "OSCAR_WILDE_Il_fantasma_di_Canterville.txt"
listLine = []
text = open(fileName)

with open(fileName, 'r') as file:
    listLine = file.readlines() #con righe vuote
    listLine = [line for line in listLine if len(line) > 1]

print(f"Il numero totale di righe nel file è: {len(listLine)}")
```

```python
charNum = 0
for line in listLine:
    charNum += len(line)
print(f"Il numero dei caratteri è: {charNum}")
```

```python
lenLineMax = ""
for line in listLine:
    if len(line) > len(lenLineMax):
        lenLineMax = line
print(f"La riga più lunga è: {lenLineMax}")
```

```python
for line in listLine:
    print(line[0])
```

```python
dictionary = {}
for line in listLine:
    for char in line:
        if char not in dictionary:
            dictionary[char] = 1
        else:
            dictionary[char] += 1
print(f"Il dizionario contiene: {dictionary}")
```

```python
text.close()
```
