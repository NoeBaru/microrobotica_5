## Esercizio

Dato un array di numeri, usare *numpy* per determinare se l'array è monotono crescente (che ogni valore è maggiore del precedente)

```python
array = np.array([1,4,6,7,8,11,12,17])
array
```

```python
np.roll(array, -1)
```

```python
np.roll(array, -1) > array
```

```python
if np.count_nonzero((np.roll(array, -1) > array)[:-1]) == (len(array)-1):
    print("L'array è monotono")
else:
    print("L'array non è monotono")
```
