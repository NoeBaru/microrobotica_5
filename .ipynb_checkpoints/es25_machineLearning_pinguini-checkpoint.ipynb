### machine learning con pinguini

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("csv/penguins_size.csv")
data.head()
```

```python
data.info()
```

**Obiettivo:**
Creare un algoritmo capace di imparare a classificare le diverse specie di pinguini a partire dalle caratteristiche di un pinguino, usando il machine learning supervisionato.

La variabile **target** è la specie presente nella colonna `species`.


Questa variabile è tipo qualitativo quindi l'algoritmo è un classificatore

## Preparazione e pulizia dei dati

```python
target = ["species"]
features = ["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]
```

```python
usefulData = data[features + target].dropna()
usefulData.head()
```

```python
usefulData.info()
```

In questa fase di preparazione dobbiamo anche trasformare le features qualitative in feature numeriche. In questo caso non è necessario

Eseguiamo l'analisi grafica dei dati

```python
np.unique(usefulData[target])
```

```python
speciesColor = {"Adelie": "Red", "Chinstrap": "Yellow", "Gentoo": "Blue"}
```

```python
fig, axs = plt.subplots(nrows = 4, ncols = 4, figsize = (15, 15))
for i, fI in enumerate(features):
    for j,fJ in enumerate(features):
        if i != j:
            for sp in speciesColor:
                filt = usefulData["species"] == sp
                axs[i, j].scatter(usefulData[filt][fI], usefulData[filt][fJ], s= 1, c = speciesColor[sp])
            axs[i, j].set_xlabel(fI)
            axs[i, j].set_ylabel(fJ)
        else:
            for sp in speciesColor:
                filt = usefulData["species"] == sp
                axs[i, j].hist(usefulData[filt][fI], edgecolor = "k", bins = 15, color = speciesColor[sp], alpha = 0.4)
            axs[i, j].set_xlabel(fI)
            axs[i, j].set_ylabel("N")
plt.show()
```

## Suddivisione dei dati in train set e test set

```python
from sklearn.model_selection import train_test_split
```

```python
X = usefulData[features] #maiuscola perché non è un solo numero, ma n (4 in questo caso)
X
```

```python
y = usefulData[target]
y
```

```python
XTrain, XTest, yTrain, yTest = train_test_split(X, y, train_size = 0.7, random_state = 123) #0.7 = 70%dati, 123 come SID, così tutti abbiamo la stessa casuale
```

# Rifacciamo parte della procedura di addestramento partendo però dalla normalizzazione delle feature che prima non abbiamo eseguito (lo abbiamo fatto nell'es26)

```python
X = usefulData[features]
y = usefulData[target]
```

Normalizziamo `X`

```python
from sklearn.preprocessing  import StandardScaler, LabelEncoder
```

```python
featureScaler = StandardScaler()
X_norm = featureScaler.fit_transform(X) #qui stiamo operando su un DataFrame
X_norm
```

```python
targetEncoder = LabelEncoder()
y_enc = targetEncoder.fit_transform(y.to_numpy().reshape((-1, 1)).ravel()) #tutte ste mille cose per togliere i warning semplicemente)
y_enc
```

## Suddivisione dei dati di training set e test set

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, yTrain, yTest = train_test_split(X_norm, y_enc, test_size = 0.3, random_state = 42)
```

## Scelta del modello

```python
from sklearn.neural_network import MLPClassifier
```

```python
model = MLPClassifier(hidden_layer_sizes = (10, 10, 6), random_state = 987, max_iter = 2000) #creo il modello della rete neurale
```

## Addestramento della rete neurale

```python
model.fit(Xtrain, yTrain)
```

## Valutazione del modello

```python
#accuratezza sui dati di test
model.score(Xtest, yTest)
```

```python
model.score(Xtrain, yTrain)
```

```python
pinguinoTest = [[38, 15.6, 205.0, 3700.0]] #mondo feature reali
specieNum = model.predict(featureScaler.transform(pinguinoTest)) #capace a prevedere i pinguini data la specie
specieNum
```

```python
targetEncoder.inverse_transform(specieNum)
```

```python
#rete neurale
iters = np.arange(10, 1000, 50)
trainAcc = []
testAcc = []
for i in iters:
    model = MLPClassifier(hidden_layer_sizes = (3, 7, 4), random_state = 987, max_iter = i, solver = "adam") #tipo di algoritmo adam
    model.fit(XTrain, yTrain)
    trainAcc.append(model.score(XTrain, yTrain))
    testAcc.append(model.score(XTest, yTest))
```

```python
fig, ax = plt.subplots(figsize = (4,4))
ax.plot(iters, trainAcc, label = "accuratezza di training")
ax.plot(iters, testAcc, label = "accuratezza di test")
ax.legend()
plt.show()
```

```python
#trovare la rete neurale più semplice che dia accuratezza più valida. Trovare la rete neurale più piccola che permetta di trovare l'accuratezza minima di 0.96
```
