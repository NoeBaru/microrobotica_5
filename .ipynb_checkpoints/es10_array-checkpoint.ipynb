### Grafici di funzioni (anticipazione su Matplotlip)

```python
x = np.linspace(-5., 5., 1000) #1000 valori equispaziati tra -5 e 5
y = x**2
fig, ax = plt.subplots(figsize=(4,2)) #due oggetti: la figura/foglio in pollici(2.54cm circa), gli assi
ax.plot(x, y)
ax.set_xlabel("x") #etichetta asse x
ax.set_ylabel("y") #etichetta asse y
ax.set_title("Questo è il titolo del grafico", fontsize = 10) #titolo
ax.grid() #griglia
plt.show()
```

```python
x = np.linspace(-5., 5., 1001) #1000 valori equispaziati tra -5 e 5
y1 = x**2
y2 = 1/x

fig, ax = plt.subplots(figsize=(4,2)) #due oggetti: la figura/foglio in pollici(2.54cm circa), gli assi

#grafico 1, label che mette 1 come pedice e il 2 come apice, colore, trasparenza colore, spessore linea, stile linea tratto-punto
ax.plot(x, y1, label = "$y_1 = x^2$", color = "blue", alpha = 0.5, linewidth = 2, linestyle = "-.")
#grafico 2 con label che mette il 2 come pedice e 1 e x come frazione, colore, spessore linea, stile della linea tratteggiata
ax.plot(x, y2, label = "$y_2 = \\frac{1}{x}$", color = "violet", linewidth = 3, linestyle = "--")
#estetica del grafico:
ax.set_xlabel("x") #etichetta asse x
ax.set_ylabel("y") #etichetta asse y
ax.set_title("Questo è il titolo del grafico", fontsize = 10) #titolo
ax.grid() #griglia
ax.set_ylim([-10,10])
ax.legend() #legenda con le label scelte durante il plot

plt.show()
```
