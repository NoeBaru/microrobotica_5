```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
# from sklearn.linear_model import LinearRegression
# from sklearn.metrics import r2_score
#from sklearn.preprocessing import  LabelEncoder
```

```python
data = pd.read_csv("csv/mnist_784_csv.csv")
data.head()
```

```python
data.info()
```

```python
features = list(data.columns)[:-1]
target = ["class"]
```

```python
X = data[features].to_numpy()
y = data[target].to_numpy()
```

```python
X.shape
```

```python
y.shape
```

```python
id_digit = 1234
fig, ax = plt.subplots(figsize = (3, 3))
ax.imshow(X[id_digit].reshape((28,28)), cmap =  "Greys")
ax.set_title(f"CIfra: {y[id_digit, 0]}")
ax.set_axis_off()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 42)
```

```python
X_train.shape
```

```python
model = MLPClassifier (hidden_layer_sizes = (10, 10), max_iter = 100) # MLPClassifier? per vedere tutte le info
```

```python
model.fit(X_train, y_train)
```

```python
print(f"Accuratezza di training: {model.score(X_train, y_train):2f}")
print(f"Accuratezza di test: {model.score(X_test, y_test):2f}")
```

```python
X_test.shape
```

```python
id_digit = 1234
digit = model.predict([X_test[id_digit]])
fig, ax = plt.subplots(figsize = (2, 2))
ax.imshow(X_test[id_digit].reshape((28,28)), cmap =  "Greys")
ax.set_title(f"Cifra: {y_test[id_digit, 0]} - La rete neurale predice: {digit[0]}")
ax.set_axis_off()
plt.show()
```

```python
 model.predict_proba([X_test[id_digit]])
```

```python
classes = model.classes_
probabilities = model.predict_proba([X_test[id_digit]])[0]
for c, p in zip(classes, probabilities):
    print(f"La cifra Ã¨ {c} con probabilitÃ  {p:.4f}")
```
