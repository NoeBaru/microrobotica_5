```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("csv/penguins_size.csv")
data.head()
```

```python
target = ["species"]
features = ["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]
```

```python
useful_data = data[target+features].dropna()
useful_data.head()
```

```python
X = useful_data[features]
y = useful_data[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
```

```python
albero = tree.DecisionTreeClassifier(max_depth=5)
```

```python
albero.fit(X_train, y_train)
```

```python
albero.score(X_test, y_test)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
tree.plot_tree(albero, feature_names=features, class_names=albero.classes_, filled=True, label="all", rounded=True)
plt.show()
```

```python
#ricerca della profondità migliore per l'albero
diz = {}
for i in range(1,15):
    albero = tree.DecisionTreeClassifier(max_depth=i)
    albero.fit(X_train, y_train)
    diz[i] = [albero.score(X_train, y_train), albero.score(X_test, y_test)]
    fig, ax = plt.subplots(figsize=(10,10))
    tree.plot_tree(albero, feature_names=features, class_names=albero.classes_, filled=True, label="all")
    plt.show()
```

```python
diz
```

```python
fig, ax = plt.subplots(figsize=(4,2))
ax.plot(list(diz.keys()), [val[0] for val in diz.values()])
ax.plot(list(diz.keys()), [val[1] for val in diz.values()])
ax.set_title("Accuratezza vs profondità")
ax.set_xlabel("Profondità albero decisionale")
ax.set_ylabel("Accuratezza")
```
