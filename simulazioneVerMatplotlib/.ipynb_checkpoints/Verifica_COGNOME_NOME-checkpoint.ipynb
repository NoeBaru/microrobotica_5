### 1

```python
dati = pd.read_csv("./pokemon.csv")
dati.head()
```

```python
dati.dropna(inplace=True) #NaN in pandas intica valori mancanti
dati.head()
```

```python
cols = ["Type1"]
dati[cols]
```

```python
bins = len(np.unique(dati[cols]))

fig, ax = plt.subplots(figsize=(6,4))

ax.hist(dati[cols].values, bins=bins, edgecolor="k")
ax.set_xticks(ax.get_xticks(), ax.get_xticklabels(), rotation=45, ha='right')
ax.set_xlabel("tipo")
ax.set_ylabel("frequenza")

plt.show()
```

### 2

```python
cols = ["Name", "Type1"]
nomi_tipi_pokemon = dati[cols].values
nomi_tipi_pokemon
```

```python
diz_pokemon = {}

for el in nomi_tipi_pokemon:
    if el[1] not in diz_pokemon:
        diz_pokemon[el[1]] = [el[0]]
    else:
        diz_pokemon[el[1]].append(el[0])

diz_pokemon
```

```python
pokemon_random = []
imgs = []
for i, tipo in enumerate(diz_pokemon):
    pokemon_random.append(np.random.choice(diz_pokemon[tipo], replace=False))
    imgs.append(plt.imread(f"./images/{pokemon_random[i]}.png"))

pokemon_random
```

```python
fig, axs = plt.subplots(nrows=2, ncols=9, figsize=(15,7))
for i, ax in enumerate(axs.reshape(-1)):
    ax.set_axis_off()

    ax.imshow(imgs[i])
    ax.set_title(pokemon_random[i])

plt.show()
```
