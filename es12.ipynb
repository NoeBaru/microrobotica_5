grafici su matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

esempio di grafico a linea

```python
x = np.linspace(-5, 5, 1000)
y = np.exp(-x**2)

x_t = np.linspace(-7, 7, 1000)
y_t = np.exp(-x_t**2)

fig, ax = plt.subplots(figsize =( 6, 4))
fig.set_facecolor("white")
ax.set_facecolor("#F5F5DC")
ax.plot(x_t, y_t, linestyle="--", linewidth = 2, color = "#CE5E6F", label=None)
ax.plot(x, y, linewidth = 2, color = "#CE5E6F", label = "$e^{-x^2}$") #oppure color = np.array([0.5, 0.1, 0.4, 0.4])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Il grafico di una curva Gaussiana", fontsize = 10)
ax.legend()
ax.grid()
plt.show()
```

```python
nPunti = 100
x = np.random.uniform(0, 1, nPunti)
y = np.random.uniform(0, 1, nPunti)

x_1 = np.random.uniform(0, 1, nPunti)
y_1 = np.random.uniform(0, 1, nPunti)

fig, ax = plt.subplots(figsize = (5, 5))
ax.set_facecolor("#1B1463")
ax.plot(x, y, linestyle = " ", marker = "o", color = "#F7F2EA", markersize = 5) #conta l'ordine di scrittura, se metto la riga dopo, i pallini vanno in secondo piano
ax.plot(x_1, y_1, linestyle = " ", marker = "X", color = "#4865ff", markersize = 5)

plt.show()
```
