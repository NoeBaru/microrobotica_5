### REGRESSIONI LINEARI

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

Proviamo a fare una regressione lineare su un dataset con una sola feature.

- feature $x$
- variabile target $y$

A partire dai dati troviamo la retta migliore che li approssima: $y = ax +b$

```python
N = 100
x = np.random.uniform(-5, 5, N)
y = 3*x - 8 + np.random.normal(0, 5, N)
```

```python
fig, ax = plt.subplots(figsize = (6, 4))
ax.scatter(x, y, color = "r", s = 10)
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
data = pd.DataFrame({"x": x, "y": y}) #da 2 array genera un dataframe
```

```python
train_test_split
```

#### DIVISIONE IN TRAIN TEST

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

#### SCELTA DEL MODELLO

```python
model = LinearRegression(fit_intercept = True)
```

#### ADDESTRAMENTO

```python
model.fit(x_train.reshape(-1, 1), y_train) #serve un array 2d altrimenti da errore -> quindi si fa reshape
```

#### VALUTAZIONE MODELLO

il metodo `.score()` per le regressioni lineari NON fornisce l'accuratezza, ma il valore di $R^2$

```python
model.score(x_test.reshape(-1, 1), y_test)
```

```python
coef_angolare = model.coef_[0]
intercetta = model.intercept_
```

```python
ascissa = np.linspace(-5, 5, 100)
ordinata = coef_angolare * ascissa  + intercetta
fig, ax = plt.subplots(figsize = (6, 4))
ax.scatter(x, y, color = "r", s = 10)
ax.plot(ascissa, ordinata)
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```
